# expanding()

parameters
1) min_periods= specifies the minimum number of observation required to have null result. if the number of observations with in the expanding window is less than min_periods;the result is NaN

2) axis=(row/columns)

```python
df1=pd.DataFrame(
    data=np.random.randint(
        low=1,
        high=5,
        size=(10,3)      
    ),
    columns=["col_1","col_2","col_3"]
)
```

```python
df1
```

```python
df1["col_1"].expanding().sum()
```

```python
df1["col_1"].expanding().mean()
```

```python
df1["col_1"].expanding().max()
```

```python
df1.expanding().sum()
```

```python
df1.expanding(
    min_periods=1,
    axis=0,
).sum()
```

```python
df1.expanding(
    min_periods=1,
    axis=1
).sum()
```

```python
df1
```

```python
df1.expanding(
    min_periods=2,
    axis=0
).sum(numeric_only=True)
```

```python
df1.expanding(
    min_periods=2,
    axis=1
).sum()
```
